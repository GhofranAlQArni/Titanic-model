# Import Libraries

let's import some libraries to get started

```python
# import python libaries 

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
```

# The Data

Let's start by reading in the titanic_train.csv file into a pandas dataframe.

```python
# import csv file 
df = pd.read_csv("titanics.csv")
```

```python
df
```

```python
df.shape
```

# Missing Data 

```python
df.info()
```

```python
df.columns
```

```python
# check for null values

pd.isnull(df).sum()
```

```python
# drop null values 

df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# machine learning algorithm does not able to undersatnd the strings that's why we convert into int. 

from sklearn.preprocessing import LabelEncoder 
```

```python
Embarked=LabelEncoder()
Sex=LabelEncoder()
df['Embarked']=Embarked.fit_transform(df['Embarked'])
df['Sex']=Sex.fit_transform(df['Sex'])

# Male = 1 , Female = 0
```

```python
df.head(5)
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)

df.describe()
```

```python
df.drop(["PassengerId","SibSp","Name","Parch","Ticket","Cabin"],axis = 1,inplace = True)
```

```python
df
```

# Exploratory Data Analysis



```python
# EDA
# plotting a countplot for survived.

import matplotlib.pyplot as plt
sns.set_style('whitegrid')
sns.countplot(x="Survived",data=df)
plt.show()
```

# Heatmap

```python
# plotting a heatmap 
x=df.head(10)
sns.heatmap(x)
```

# Sex

```python
# plotting a bar chart for survived vs sex

sns.set(style="darkgrid")
sns.barplot(x="Sex",y="Survived",data=df)
plt.show()
```

# Age

```python
# plotting a barplot for survived vs age

sns.set(style="darkgrid")
bins = pd.cut(df['Age'],bins=4) #create 4  bins
df['Age_binned'] = bins
sns.barplot(x="Age_binned",y="Survived",data=df,orient="v")
plt.show()
```

# Fare

```python
# plotting a barplot for survived vs Fare

sns.set(style="darkgrid")
bins = pd.cut(df['Fare'],bins=3) #create 3  bins
df['Fare_binned'] = bins
sns.barplot(x="Fare_binned",y="Survived",data=df,orient="v")
plt.show()
```

# Embarked

```python
# plotting a bar plot for survived vs embarked

sns.set(style="darkgrid")
bins = pd.cut(df['Embarked'],bins=4) #create 4  bins
df['Embarked_binned'] = bins
sns.barplot(x="Embarked_binned",y="Survived",data=df,orient="v")
plt.show()
```

# Pclass

```python
# plotting a bar plot for survived vs pclass

sns.set(style="darkgrid")
bins = pd.cut(df['Pclass'],bins=5) #create 5  bins
df['Pclass_binned'] = bins
sns.barplot(x="Pclass_binned",y="Survived",data=df,orient="v")
plt.show()
```

# Pairplot

```python
# plotting the pairplot
sns.pairplot(df)
plt.show()
```

# Logistic regression with python 

```python
# Using logistic regression for predicting the model.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['Pclass','Age','Fare','Embarked']],df[['Survived']],test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression 

lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict([[0,56.0,83.1583,0]])
```

```python
lr.predict([[1,54.0,51.8625,2]])
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict(x_test)
```
